```python
# creating odd list of K for KNN
mylist = list(range(1,50,2))
```

```python
# empty list that will hold accuracy scores
ac_scores = []
ac_scores1 = []


# perform accuracy metrics for values from 1,3,5....19
for k in mylist:
    knn = MultiOutputClassifier(KNeighborsClassifier(n_neighbors=k))
    knn.fit(X_train.to_numpy(), y_train.to_numpy())
    # predict the response
    y_pred = knn.predict(X_val.to_numpy())
    # evaluate accuracy
    scores = accuracy_score(y_val, y_pred)
    ac_scores.append(scores)
    scores1 = average_precision_score(y_val.to_numpy(),pred,average='samples')
    ac_scores1.append(scores1)

# changing to misclassification error
MSE = [1 - x for x in ac_scores]

# determining best k
optimal_k = mylist[MSE.index(min(MSE))]
print("The optimal number of neighbors is %d" % optimal_k)
```
